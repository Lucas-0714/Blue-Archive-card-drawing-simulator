```python
from IPython.display import Image, display
from IPython.display import HTML, display
from IPython.display import clear_output
import time
import random

def displayimage(link, delay):
  # exchange by display(Image(url=link))
  display(HTML(f'<img src="{link}" alt="animation" width="300px">'))
  time.sleep(delay)

def run():
  displayimage("https://i.imgur.com/Y2aak8Y.jpeg", 0.3)
  clear_output(wait=True)
  displayimage("https://i.imgur.com/BPLJ8fD.jpeg", 0.3)
  clear_output(wait=True)
  displayimage("https://i.imgur.com/v7igECD.jpeg", 0.3)
  clear_output(wait=True)
  displayimage("https://i.imgur.com/46zPj7I.jpeg", 0.3)
  clear_output(wait=True)
  displayimage("https://i.imgur.com/ziHJyHO.jpeg", 0.3)
  clear_output(wait=True)
  displayimage("https://i.imgur.com/m2C4wXk.jpeg", 0.3)
  clear_output(wait=True)
  displayimage("https://i.imgur.com/6ERPNHF.jpeg", 0.3)
  clear_output(wait=True)
  displayimage("https://i.imgur.com/wyrXRFK.jpeg", 0.3)
  clear_output(wait=True)
  displayimage("https://i.imgur.com/2aJnQq9.jpeg", 0.3)
  clear_output(wait=True)
  displayimage("https://i.imgur.com/7tTSOYB.jpeg", 0.3)
  clear_output(wait=True)
  displayimage("https://i.imgur.com/16qx9a4.jpeg", 0.3)
  clear_output(wait=True)
  displayimage("https://i.imgur.com/Oa6hrcn.jpeg", 0.3)
  clear_output(wait=True)
  displayimage("https://i.imgur.com/s6v35Wg.jpeg", 0.3)
  clear_output(wait=True)
  displayimage("https://i.imgur.com/flqyJ4T.jpeg", 0.3)
  clear_output(wait=True)
  displayimage("https://i.imgur.com/gbmsxqw.jpeg", 0.3)
  clear_output(wait=True)
  displayimage("https://i.imgur.com/UUue8mz.jpeg", 0.3)

def speed():
  displayimage("https://i.imgur.com/Y2aak8Y.jpeg", 0)
  displayimage("https://i.imgur.com/BPLJ8fD.jpeg", 0)
  displayimage("https://i.imgur.com/v7igECD.jpeg", 0)
  displayimage("https://i.imgur.com/46zPj7I.jpeg", 0)
  displayimage("https://i.imgur.com/ziHJyHO.jpeg", 0)
  displayimage("https://i.imgur.com/m2C4wXk.jpeg", 0)
  displayimage("https://i.imgur.com/6ERPNHF.jpeg", 0)
  displayimage("https://i.imgur.com/wyrXRFK.jpeg", 0)
  displayimage("https://i.imgur.com/2aJnQq9.jpeg", 0)
  displayimage("https://i.imgur.com/7tTSOYB.jpeg", 0)
  displayimage("https://i.imgur.com/16qx9a4.jpeg", 0)
  displayimage("https://i.imgur.com/Oa6hrcn.jpeg", 0)
  displayimage("https://i.imgur.com/s6v35Wg.jpeg", 0)
  displayimage("https://i.imgur.com/flqyJ4T.jpeg", 0)
  displayimage("https://i.imgur.com/gbmsxqw.jpeg", 0)
  displayimage("https://i.imgur.com/UUue8mz.jpeg", 0)
  displayimage("https://i.imgur.com/dG66JmO.png", 0)
  displayimage("https://i.imgur.com/x2l8bjH.jpeg", 0)
  time.sleep(2.5)
  clear_output(wait=True)
  time.sleep(2.5)

global li
li = []
def card():
  cardlink = ["https://i.imgur.com/MEznrP3.jpeg", "https://i.imgur.com/2caFmiG.jpeg", "https://i.imgur.com/duHYFz4.jpeg"]
  # link: blue, golden, rainbow
  global rand
  rand = random.randint(0, 1000)
  link = ""
  if rand >= 0 and rand <= 30:
    link = cardlink[2]
  elif rand >= 31 and rand <= 225:
    link = cardlink[1]
  elif rand >= 256 and rand <= 1000:
    link = cardlink[0]
  else:
    print(f'Error')
  global li
  li.append(link)

def home():
  clear_output(wait=True)
  global stone

  # 初始化青輝石
  global stone
  stone = random.randint(0,100000) + 1200
  stone = stone + 1200
  display(HTML(f'''
  <div style="display: flex;">
    <img src="https://i.imgur.com/dG66JmO.png" alt="stone" height="25px" width="25px">
    <text> × {stone}</text>
  </div>
  '''))
  stone = stone - 1200
  displayimage("https://i.imgur.com/x2l8bjH.jpeg", 0)

#  • start

# 開始
stone = 0
def start():
  # 動畫
  clear_output(wait=True)
  run()

  # 準備抽卡
  clear_output(wait=True)
  time.sleep(0.4)

  # 隨機抽卡
  for i in range(10):
    card()
    global rand
    print(rand)
  global li
  v1c = li[0]
  v2c = li[1]
  v3c = li[2]
  v4c = li[3]
  v5c = li[4]
  # ---------
  v6c = li[5]
  v7c = li[6]
  v8c = li[7]
  v9c = li[8]
  v10c = li[9]
  clear_output(wait=True)

  # 顯示青輝石
  display(HTML(f'''
  <div style="display: flex;">
    <img src="https://i.imgur.com/dG66JmO.png" alt="stone" height="25px" width="25px">
    <text> × {stone}</text>
  </div>
  '''))
  # 顯示卡片
  display(HTML(f'''
  <div style="display: flex;">
    <div>
      <img src="{v1c}" alt="card 1" width="60px">
    </div>
    <div>
      <img src="{v2c}" alt="card 2" width="60px">
    </div>
    <div>
      <img src="{v3c}" alt="card 3" width="60px">
    </div>
    <div>
      <img src="{v4c}" alt="card 4" width="60px">
    </div>
    <div>
      <img src="{v5c}" alt="card 5" width="60px">
    </div>
  </div>
  <br>
  <div style="display: flex;">
    <div>
      <img src="{v6c}" alt="card 6" width="60px">
    </div>
    <div>
      <img src="{v7c}" alt="card 7" width="60px">
    </div>
    <div>
      <img src="{v8c}" alt="card 8" width="60px">
    </div>
    <div>
      <img src="{v9c}" alt="card 9" width="60px">
    </div>
    <div>
      <img src="{v10c}" alt="card 10" width="60px">
    </div>
  '''))

speed()
home()
time.sleep(7.5)
start()
```
